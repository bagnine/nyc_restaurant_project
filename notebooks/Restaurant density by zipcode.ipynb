#### data downloaded from https://data.cityofnewyork.us/Health/Restaurants-rolled-up-/59dk-tdhz

```python
import pandas as pd
import ast
```

```python
df = pd.read_csv('Restaurants__rolled_up_.csv')
```

```python
df.astype({'ZIPCODE':'object'})
```

```python
zip_values = df['ZIPCODE'].value_counts()
```

```python
zip_values.index
```

```python
num_rest_by_zip_df = pd.DataFrame(data={'zip_code':zip_values.index, 'num_rest':zip_values})
```

```python
num_rest_by_zip_df
```

```python
list(range(0,len(num_rest_by_zip_df)))
```

```python
for n in list(range(0,len(num_rest_by_zip_df))):
    num_rest_by_zip_df['index'][n] = n
```
